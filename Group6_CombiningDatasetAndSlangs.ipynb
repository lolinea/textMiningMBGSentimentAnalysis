## Preprocessing

On secondary dataset (Retrieved from Kaggle), there are 16 rows where the tweet/post are not stated in Indonesia, therefore, these rows will be dropped

```python
df2[df2['lang'] != 'in'].info()
```

```python
df2 = df2[df2['lang'] == 'in'].reset_index(drop=True)
df2.info()
```

From the secondary dataset, only three rows: id_str, username, and full_text will be taken. These rows matches with the first dataset (Retrieved from Scraping) and the rows will be renamed as tweet_id, username, and tweet_content so it can be combined.

```python
columns = ['id_str', 'username', 'full_text']
df2 = df2[columns]
df2.head()
```

```python
df2.columns = ['tweet_id', 'username', 'tweet_content']
df2.head()
```

```python
df = pd.concat([df1, df2]).reset_index()
df.info()
```

After both dataset are combined, tweet_label has no values because both datasets has no labels for the sentiment. IndoRoBERTa will be applied to label these sentiments automatically. 

```python
df[df['tweet_content'].duplicated()]
```

The duplicated datas available after combining both datasets will be dropped. 

```python
df = df.drop_duplicates(subset='tweet_content').reset_index(drop=True)
df.info()
```

Combined dataset will be saved as CSV and will be used for for modelling.

```python
df.to_csv('data.csv', index=False)
```

# Combining Slangs

```python
slangs = pd.read_csv("slang_indo.csv", header=None)
slangs = dict(zip(slangs[0], slangs[1]))
slangs
```

```python
slangs2 = open("combined_slang_words.txt")
slangs2 = json.loads(slangs2.read())
slangs2
```

```python
combined = {**slangs, **slangs2}
combined
```

There are two datasets retrieved from Kaggle containing list of slangs and their formal words. Those dataset will be loaded and combined as one list. The combined data will be saved to be used when preprocessing.

```python
with open('slangs.json', 'w') as file:
    json.dump(combined, file)
```
